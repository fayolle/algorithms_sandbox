Algorithm 7.2.1.5-H for enumerating partitions using restricted growth strings

```python
import copy
```

```python
def partitions(n):
  if n<2:
    print('Input n should be >= to 2')
    return

  # Initialize
  a = [0] * n
  b = [1] * (n-1)
  m = 1

  ps = []

  while True:
    ps.append(copy.copy(a))

    if a[n-1] != m:
      a[n-1] = a[n-1] + 1
      continue

    j = n-2
    while a[j] == b[j]:
      j = j - 1

    if j == 0:
      return ps

    a[j] = a[j] + 1

    if a[j] == b[j]: 
      m = b[j] + 1
    else:
      m = b[j]

    j = j + 1

    while j < n-1:
      a[j] = 0
      b[j] = m
      j = j + 1 

    a[n-1] = 0
```

```python
n = 4
ps = partitions(n)
for p in ps:
  print(p)
```

```python
def pretty_print(p):
  a = {}
  n = len(p)
  for i in range(n):
    c = p[i]
    if not(c in a):
      a[c] = [i]
    else:
      a[c].append(i)

  res = ''
  for k in a:
    for e in a[k]:
      res = res + ' '.join(str(e))
    res = res + '|'

  return res[:-1]
```

```python
n = 4
ps = partitions(n)
for p in ps:
  print(pretty_print(p))
```
